```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
df = pd.read_csv('dados-site.csv', sep=';', decimal=',')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
x = df['ano']
y1 = df['criancas']
y2 = df['variacao_criancas(%)']

fig, ax1 = plt.subplots()
bars = ax1.bar(x, y1)
plt.bar_label(bars, label_type='edge')

ax1.set_ylabel('Quantidade de alunos')
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red')
ax2.set_ylabel('Variação (%)')

plt.title('Quantidade de alunos e Variação por ano')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

plt.show()
```

```python
x = df['ano']
y1 = df['bolsitas_escola_parceira']
y2 = df['variacao_bolsitas_escolas_parceiras(%)']

fig, ax1 = plt.subplots()
bars = ax1.bar(x, y1)
plt.bar_label(bars, label_type='edge')

ax1.set_ylabel('Quantidade de balsas')
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red')
ax2.set_ylabel('Variação (%)')

plt.title('Evolução do número de bolsas obtidas por ano')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

plt.show()
```
